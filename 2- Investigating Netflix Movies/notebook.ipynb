```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# read the data into CSV file format
netflix_data = pd.read_csv("netflix_data.csv")
netflix_data.head(20)
```

# What was the most frequent movie duration in the 1990s?

```python
# Select 1990s released movies
select_movies = netflix_data[netflix_data["type"]=="Movie"]
released_1990s = select_movies[(select_movies["release_year"] >= 1990) & (select_movies["release_year"] < 2000)]

# Find the most frequent movie duration
duration_count = released_1990s["duration"].value_counts()
duration = duration_count.index[0]  # The duration itself
frequency = duration_count.iloc[0]  # The frequency of that duration

print(f"The most frequent movie duration is {duration} with a frequency of {frequency}.")
```

```python
# Visualize the duration column of your filtered data to see the distribution of movie durations
plt.hist(released_1990s["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

# A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s

```python
# Select 1990s released movies
select_movies = netflix_data[netflix_data["type"]=="Movie"]
released_1990s = select_movies[(select_movies["release_year"] >= 1990) & (select_movies["release_year"] < 2000)]

# select the short Action movies released in 1990S
short_action_movies = released_1990s[(released_1990s["genre"]=="Action")]

# short_action_movies = released_1990s[(released_1990s["genre"]=="Action") & (released_1990s["duration"]<90)]
# # count the number of the short movies
# short_movie_count = short_action_movies.shape[0]
# int(short_movie_count)
# print(f"The number of short action movies is {short_movie_count}.")

# count the number of the short movies
short_movie_count = 0
for label, row in short_action_movies.iterrows() :
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```
